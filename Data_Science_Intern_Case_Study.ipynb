# Side Effect Frequency

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_histogram(df):
   
    plt.figure(figsize=(10, 6))
    yan_etki_counts = df['Yan_Etki'].value_counts()
    
    yan_etki_counts.plot(
        kind='bar', color='blue', edgecolor='black'
    )
    
    plt.title('Yan Etki Frekansı')
    plt.xlabel('Yan Etkiler')
    plt.ylabel('Frekans')
    plt.xticks(rotation=90)
    plt.show()

plot_histogram(df)
```

# Side Effect Status by Gender

```python

import matplotlib.pyplot as plt

yan_etki_cinsiyet = (
    df.groupby('Cinsiyet')['Yan_Etki']
    .value_counts(normalize=True)
    .unstack()
)
print(yan_etki_cinsiyet)

yan_etki_cinsiyet = (
    df.groupby(["Cinsiyet", "Yan_Etki"])
    .size()
    .unstack()
    .fillna(0)
)
yan_etki_cinsiyet.plot(kind="bar", stacked=True, figsize=(10, 7))
plt.title("Cinsiyete Göre Yan Etki Dağılımı")
plt.show()
```

# Side Effect Status According to Medicines

```python
import matplotlib.pyplot as plt
import seaborn as sns
yan_etkiler = (
    df.groupby(["Ilac_Adi", "Yan_Etki"])
    .size()
    .unstack()
    .fillna(0)
)
plt.figure(figsize=(12, 8))
sns.heatmap(yan_etkiler, cmap="YlGnBu", linewidths=0.5)

plt.title("İlaçlardan Alınan Yan Etkiler")
plt.ylabel("İlaç Adı")
plt.xlabel("Yan Etki")

plt.tight_layout()
plt.show()

```

```python
def plot_drug_side_effects(df):
  
    plt.figure(figsize=(10, 6))
    
    sns.countplot(
        y='Ilac_Adi', hue='Yan_Etki', data=df, palette='Set1',
        order=df['Ilac_Adi'].value_counts().index
    )
    
    plt.title('İlaç Başına Yan Etki Dağılımı')
    plt.xlabel('Yan Etki Sayısı')
    plt.ylabel('İlaç Adı')
    plt.legend(title='Yan Etki', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

plot_drug_side_effects(df)
```

```python
import matplotlib.pyplot as plt
df.groupby("Ilac_Adi")["Yan_Etki"].count().plot(figsize=(10, 6))

plt.title("İlaçtan Alınan Yan Etki Miktarı")
plt.xticks(rotation=45, ha='right')  
plt.text(5, 35, "İLAÇTAN ALINAN YAN ETKİ SIKLIĞI", 
         bbox=dict(facecolor="yellow", alpha=0.5))
plt.tight_layout() 
plt.show()

```

# Filling Missing Values

```python
import pandas as pd
from sklearn.impute import KNNImputer


df['Cinsiyet'] = df['Cinsiyet'].fillna(df['Cinsiyet'].mode()[0])

df['Il'] = df['Il'].fillna(df['Il'].mode()[0])

df['Alerjilerim'] = df['Alerjilerim'].fillna('Bilinmiyor')

df['Kronik Hastaliklarim'] = df['Kronik Hastaliklarim'].fillna('Bilinmiyor')

df['Baba Kronik Hastaliklari'] = df['Baba Kronik Hastaliklari'].fillna('Bilinmiyor')

df['Anne Kronik Hastaliklari'] = df['Anne Kronik Hastaliklari'].fillna('Bilinmiyor')

df['Kiz Kardes Kronik Hastaliklari'] = df['Kiz Kardes Kronik Hastaliklari'].fillna('Bilinmiyor')

df['Erkek Kardes Kronik Hastaliklari'] = df['Erkek Kardes Kronik Hastaliklari'].fillna('Bilinmiyor')

df['Kan Grubu'] = df['Kan Grubu'].fillna('Bilinmiyor')

df['Kilo'] = df['Kilo'].fillna(df['Kilo'].median())

df['Boy'] = df['Boy'].fillna(df['Boy'].median())
```
